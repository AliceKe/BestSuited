```python
import json
f = open('../init.json')
data = json.load(f)
```

```python
data
```

```python
jobs = set()
locations = set()
```

```python
for posting in data['job_postings']:
  jobs.add(posting['role'])
  locations.add(f"{posting['city']},{posting['country']}")
```

```python
jobsSet = []
for i in jobs:
  jobsSet.append({"title": i})
print(jobsSet)
```

```python
locationsSet = []
for i in locations:
  locationsSet.append({"title": i})
print(locationsSet)
```
